## Choisir un mot

```python
mot = input('Quel mot souhaitez-vous ? ')
```

## Scraping

```python
from requests_html import HTMLSession
```

```python
session = HTMLSession()
url = "https://www.ldoceonline.com/dictionary/"
urlComplet = url + mot
```

```python
response = session.get(urlComplet)
html_doc = response.content
```

## Parsing

```python
from bs4 import BeautifulSoup
```

```python
soup = BeautifulSoup(response.text,'html.parser')
soup
```

```python
souplettesBody = soup.find('body')
souplettesBody
```

```python
finded = souplettesBody.findAll('div',{"class":"dictionary"})
```

## Lien UK

```python
for v in finded:
    baliseAnglais = v.findAll('span',{"class":"speaker brefile fas fa-volume-up hideOnAmp"})
```

```python
for f in baliseAnglais:
    liensAnglais = f['data-src-mp3']
```

```python
liensAnglais = liensAnglais.replace(liensAnglais[-15:],"")
```

## Lien US

```python
for v in finded:
    baliseUS = v.findAll('span',{"class":"speaker amefile fas fa-volume-up hideOnAmp"})
```

```python
for f in baliseUS:
    liensUS = f['data-src-mp3']
    print(liensUS)
```

```python
liensUS = liensUS.replace(liensUS[-15:],"")
```

## Le mot 

```python
for v in finded:
    motVague = v.find('span',{"class":"HYPHENATION"})
mot = motVague.text.strip()
```

## Catégorie 

```python
for v in finded:
    categVague = v.find('span',{"class":"POS"})
categ = categVague.text.strip()
```

## Prononciation

```python
for v in finded:
    prononVague = v.find('span',{"class":"PRON"})
prononciation = prononVague.text.strip()
```

## Résultat 

```python
print("Le mot est : " + mot + ".")
print("Il appartient à la catégorie : " + categ + ".")
print("Il se prononce : " + prononciation + ".")
print("Prononciation anglaise : " + liensAnglais)
print("Prononciation américaine : " + liensUS)
```
