# Les données

```python
cheminDico = "./Data/1757_Buchanan-J.csv" 

dfDicoBuch = pd.read_csv(cheminDico, delimiter=";", encoding="utf8")

# afficher à l'écran un échantillon du tableau
dfDicoBuch.sample(n=5)
```

* Table mapping

```python
cheminMap = "./Data/Buchanan_PronChar_counts.csv"
dfMap = pd.read_csv(cheminMap, delimiter=";", encoding="utf8")

# Affichage
dfMap.sample(n=5)
```

On remplace les vides par des chaines de caractères vides

```python
dfMap = dfMap.replace(numpy.nan, '', regex=True)
```

# Fonction qui change le texte

```python
def changeTexte(motMoche,dfMap):
    Resultat = ""
    for lettre in motMoche: 
        dfRowFinded = dfMap[dfMap['char']== lettre]
        if len(dfRowFinded==1):
            Resultat+=dfRowFinded.iloc[0]['graph']
        else:
            Resultat+='?'
    return(Resultat)
```

# Nouvelle colonne motVedette

```python
dfDicoBuch['motVedette'] = dfDicoBuch.apply(lambda x: changeTexte(x['pron'], dfMap),1)
```
